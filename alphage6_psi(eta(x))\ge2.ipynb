```julia
using IntervalArithmetic

function check(func,interval_check)
    lower=interval_check[1]
    upper=interval_check[2]
    beg=lower
    print(beg, "\n")
    last=upper+1
    temp=beg..last
    while beg<upper+1/(2^(30))
        res=func(temp)
        iters=0
        lim=0..0
        while !(((res ∩ lim)==∅) & (res>lim))
            last=(beg+last)/2
            temp=beg..last
            res=func(temp)
            iters=iters+1
            if iters>=20
                last=nextfloat(beg)
                #print(beg," ",last, "\n")
                iters=0
            end
        end
        print("Checked from ",lower, " up to ",beg, "\n")  
        beg=last
        last=beg+1
        temp=beg..last
    end
    print("Checked on interval ", "[",lower,",",upper,"]","\n")
end
    

    
```

```julia
alphas=[6,8,10,12,14] #1.408
lp_salphas=[7.83668,11.813148, 15.830697, 19.854857, 23.876137]
up_salphas=[7.83669, 11.813151, 15.830699, 19.854859, 23.876138]


for j in [1,2,3,4,5]
    print('\n','\n','\n')
    alpha=alphas[j]
    s_alphar=lp_salphas[j]..up_salphas[j]
    function F_a(x)
        1/(1+s_alphar*(x^alpha))
    end
    
    function dF_a(x)
        if x==0
            val=0
        else
            val=-(alpha*F_a(x)*(1-F_a(x)))/x
        end
    end
    
    function Q(x,n)
        F_a(n)/((x-n)^2)+dF_a(n)/(x-n)
    end
    
    p=30
    function quant(x)
        al=alpha..alpha
        val=8*(al+1)*(al+p-1)/(((p..p)^al)*(al-1))
        for n=1:(p-1)
            if n==round(x)
                val=val+Q(x,-n)
            else
                val=val+Q(x,n)+Q(x,-n)
            end
        end
        -(val+Q(x,0))
        #-Q(x,-3)
    end
    
     print("On alpha = ",alpha, '\n')
     print(quant(1.5),'\n')
     check(quant,[1.5,10])
 end
```
